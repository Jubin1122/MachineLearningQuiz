# Color Threshold, Blue Screen
## Import resources

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

%matplotlib inline
```

## Read in and display image

```python
image = cv2.imread('img/pizza_bluescreen.jpg')

# the image type and dimesnions
print(f'This image is: {type(image)} with dimensions: {image.shape}')
```

```python
# copy of the image
img_cpy = np.copy(image)

# change color to RGB(from BGR)
img_cpy = cv2.cvtColor(img_cpy, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(img_cpy)
```

## Define the color threshold

```python
## Define the color selection boundaries in RGB values
# play around with these values until you isolate the blue background

lower_blue = np.array([0,0,230])
upper_blue = np.array([50,50,255])
```

```python
# define the masked area
mask = cv2.inRange(img_cpy, lower_blue, upper_blue)

# vizulaize the mask
plt.imshow(mask, cmap='gray')
```

```python
# Mask the image to let the pizaa show through
masked_image =  np.copy(img_cpy)
masked_image[mask!=0] = [0,0,0]

# Display it
plt.imshow(masked_image)
```

## Mask and a background image

```python
back_img = cv2.imread('img/space_background.jpg')
back_img = cv2.cvtColor(back_img, cv2.COLOR_BGR2RGB)

back_img.shape

# Crop it to the right size (514x816)
crop_background = back_img[0:514, 0:816]

#mask the cropped background so that pizza area is blocked
crop_background[mask==0]=[0,0,0]

plt.imshow(crop_background)
```

## **Create a complete image**

```python
# Add the two images together to create a complete image
comp_img = masked_image + crop_background

# Display the result
plt.imshow(comp_img)
```
